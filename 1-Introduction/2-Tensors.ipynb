## Images as PyTorch Tensors

```python
import torch

# create a 3 dimensional tensor
#  Creates 4 images defiend by 28x28
images = torch.rand(4,28,28)

#get second image
second_image = images[1]
```

## Displaying Images

```python
import matplotlib.pyplot as plt

plt.imshow(second_image,cmap='gray')
plt.axis('off')
plt.show()
```
